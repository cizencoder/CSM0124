b) ¬p → (q → r)

c) (p → q) ∨ (¬p → r)

d) (p → q) ∧ (¬p → r)

f) (¬p ↔ ¬q) ↔ (q ↔ r)


```python
nice = []
for p in [True, False]:
    for q in [True, False]:
        for r in [True, False]:
          prop1 = not p
          prop2 = not q
          prop3 = not r
          prop4 = prop2 or r
          prop5 = q if p else True
          prop6 = r if q else True
          prop7 = r if prop1 else True
          prop8 = prop4 if p else True
          prop9 = prop6 if prop1 else True
          prop10 = prop5 or prop7
          prop11 = prop5 and prop7
          prop1210 = p if q else True
          prop121 = prop5 == prop1210 #Biimplikasi P dan Q
          prop1220 = r if prop2 else True
          prop1221 = prop2 if r else True
          prop122 = prop1220 == prop1221 #Biimplikasi not Q dan R
          prop12 = prop121 or prop122
          prop1310 = prop2 if prop1 else True
          prop1311 = prop1 if prop2 else True
          prop131 = prop1310 == prop1311
          prop1320 = q if r else True
          prop132 = prop6 == prop1320
          prop13 = prop131 == prop132
          nice.append(pd.DataFrame({'p': [p],
                                   'q': [q],
                                   'r': [r],
                                   '¬p': [prop1],
                                   '¬q':[prop2],
                                   '¬r':[prop3],
                                   '¬q ∨ r':[prop4],
                                   'p → q':[prop5],
                                   'q → r':[prop6],
                                   '¬p → r':[prop7],
                                   'p ↔ q':[prop121],
                                   '¬q ↔ r':[prop122],
                                   '¬p ↔ ¬q':[prop131],
                                   'q ↔ r':[prop132],
                                   'p → (¬q ∨ r)':[prop8],
                                   '¬p → (q → r)': [prop9],
                                   '(p → q) ∨ (¬p → r)' : [prop10],
                                   '(p → q) ∧ (¬p → r)' : [prop11],
                                   '(p ↔ q) ∨ (¬q ↔ r)' : [prop12],
                                   '(¬p ↔ ¬q) ↔ (q ↔ r)' : [prop13],
                                  }))

tableSoal1 = pd.concat(nice, ignore_index=True)
tableSoal1
```

2. Show that each of these conditional statements is a tautology by using truth tables.<br>
a) [¬p ∧ (p ∨ q)] → q <br>
b) [(p → q) ∧ (q → r)] → (p → r) <br>
c) [p ∧ (p → q)] → q <br>
d) [(p ∨ q) ∧ (p → r) ∧ (q → r)] → r <br>


```python
# a) [¬p ∧ (p ∨ q)] → q
dfs = []
tautologi = True
for p in [True, False]:
    for q in [True, False]:
        prop1 = not p
        prop2 = p or q
        prop3 = prop1 and prop2
        prop4 = q if prop3 else True
        if(not prop4):
            tautologi = False
        dfs.append(pd.DataFrame({'P': [p],
                                 'Q': [q],
                                 '¬p': [prop1],
                                 'p ∨ q':[prop2],
                                 '¬p ∧ (p ∨ q)':[prop3],
                                 '[¬p ∧ (p ∨ q)] → q': [prop4],
                                }))

tb_soal_21 = pd.concat(dfs, ignore_index=True)

if(tautologi):
    print("Merupakan Tautologi")
else:
     print("Bukan Merupakan Tautologi")
tb_soal_21
```

```python
# b) [(p → q) ∧ (q → r)] → (p → r)
dfs = []
tautologi = True
for p in [True, False]:
    for q in [True, False]:
        for r in [True, False]:
            prop1 = q if p else True
            prop2 = r if q else True
            prop3 = r if p else True
            prop4 = prop1 and prop2
            prop5 = prop3 if prop4 else True
            if(not prop5):
                tautologi = False
            dfs.append(pd.DataFrame({'P': [p],
                                    'Q': [q],
                                    'p → q': [prop1],
                                    'q → r':[prop2],
                                    'p → r':[prop3],
                                    '(p → q) ∧ (q → r)': [prop4],
                                    '[(p → q) ∧ (q → r)] → (p → r)': [prop5],
                                    }))

tb_soal_22 = pd.concat(dfs, ignore_index=True)

if(tautologi):
    print("Merupakan Tautologi")
else:
     print("Bukan Merupakan Tautologi")
tb_soal_22
```

```python
# c) [p ∧ (p → q)] → q
dfs = []
tautologi = True
for p in [True, False]:
    for q in [True, False]:
        prop1 = q if p else True
        prop2 = p and prop1
        prop3 = q if prop2 else True
        if(not prop3):
            tautologi = False
        dfs.append(pd.DataFrame({'P': [p],
                                 'Q': [q],
                                 'p → q': [prop1],
                                 'p ∧ (p → q)':[prop2],
                                 '[p ∧ (p → q)] → q':[prop3],
                                }))

tb_soal_23 = pd.concat(dfs, ignore_index=True)

if(tautologi):
    print("Merupakan Tautologi")
else:
     print("Bukan Merupakan Tautologi")
tb_soal_23
```

```python
# d) [(p ∨ q) ∧ (p → r) ∧ (q → r)] → r
dfs = []
tautologi = True
for p in [True, False]:
    for q in [True, False]:
        for r in [True, False]:
            prop1 = p or q
            prop2 = r if p else True
            prop3 = r if q else True
            prop4 = prop1 and prop2 and prop3
            prop5 = r if prop4 else True
            if(not prop5):
                tautologi = False
            dfs.append(pd.DataFrame({'P': [p],
                                    'Q': [q],
                                    'p ∨ q': [prop1],
                                    'p → r':[prop2],
                                    'q → r':[prop3],
                                    '(p ∨ q) ∧ (p → r) ∧ (q → r)':[prop4],
                                    '[(p ∨ q) ∧ (p → r) ∧ (q → r)] → r':[prop5],
                                    }))

tb_soal_24 = pd.concat(dfs, ignore_index=True)

if(tautologi):
    print("Merupakan Tautologi")
else:
     print("Bukan Merupakan Tautologi")
tb_soal_24
```

4. Seorang detektif sedang menangani sebuah kasus kriminal. Ada 4 orang tersangka berinisial
G, O, L, D. Dari hasil interogasi, detektif tersebut mendapatkan sejumlah fakta, yaitu :
- Jika G berkata jujur, maka O berkata jujur.
- Jika L berkata bohong, maka D berkata jujur.
- Salah satu dari O dan L berkata jujur.
- O berkata jujur jika dan hanya jika D berkata bohong.
Dari sejumlah fakta di atas, siapa yang berkata jujur? buatlah pembuktiannya dengan tabel kebenaran!




*   G -> O
*   ~L <-> D
*   O xor L
*   O <-> ~D





```python
tersangka = []
for g in [True, False]:
    for o in [True, False]:
        for l in [True, False]:
            for d in [True, False]:
                prop1 = not l
                prop2 = not d
                prop3 = d if prop1 else True
                prop4 = prop1 if d else True
                prop5 = prop2 if o else True
                prop6 = o if prop2 else True
                prop7 = o if g else True
                prop8 = d if prop1 else True
                prop9 = o != l
                prop10 = prop5 == prop6
                tersangka.append(pd.DataFrame({'G' : [g],
                                               'O' : [o],
                                               'L' : [l],
                                               'D' : [d],
                                               '~L' : [prop1],
                                               '~D' : [prop2],
                                               '~L -> D' : [prop3],
                                               'D -> ~L' : [prop4],
                                               'O -> ~D' : [prop5],
                                               '~D -> O' : [prop6],
                                               'G -> O' : [prop7],
                                               '~L <-> D' : [prop8],
                                               'O or L' : [prop9],
                                               'O <-> ~D' : [prop10]
                                               }))

tabel_soal_4 = pd.concat(tersangka, ignore_index=True)
tabel_soal_4
```
